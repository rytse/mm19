```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
avg = pd.read_csv('2019_avg_cps.csv')

def my_level_list(data):
    A = []
    for i in range(len(data)):
        if (data[i]*100/np.sum(data)) > 7 : #2%
            A.append(avg['Team'][i])
        else:
            A.append('')
    return A

plt.pie(avg['Round 7'], radius = 1, labels=my_level_list(avg['Round 7']), shadow=False)
plt.title('Predicted Probabilties of Winning it All')
plt.savefig('pie.png', format='png', dpi=1000)
plt.show()
```
